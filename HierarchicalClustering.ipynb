```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

**Example 1 - Diabetes Data**

```python
# Loading the diabetes dataset
data=pd.read_csv("/content/diabetes.csv")
```

```python
data.head()
```

```python
df=data.drop(['Outcome'], axis=1)
```

```python
# Importing hierarchical clustering functions from scipy
import scipy.cluster.hierarchy as sch

# Creating a dendrogram for the diabetes dataset
plt.title("Diabetes Dendogram")
dendrogram=sch.dendrogram(sch.linkage(df,method='ward'))
```

**Example 2 - Country Clusters Data**

```python
# Loading the country clusters dataset
country_data=pd.read_csv("/content/Country_clusters.csv")
```

```python
country_data.head()
```

```python
# Selecting only two columns(latitude and longitude) from the country clusters dataset
country_data=country_data.iloc[:,1:3]
```

```python
country_data.head()
```

```python
# Creating a dendrogram for the country clusters dataset
plt.title("Country Clusters dendrogram")
dendrogram=sch.dendrogram(sch.linkage(country_data,method='ward'))
```

**Example 3 - Wholesale Customers Data**

```python
# Loading the wholesale customers dataset
new_data=pd.read_csv("/content/Wholesale_customers.csv")
```

```python
new_data.head()
```

```python
# Creating a dendrogram for the wholesale customers dataset
plt.title("Wholesale dendrogram")
dendrogram=sch.dendrogram(sch.linkage(new_data,method='ward'))
```
